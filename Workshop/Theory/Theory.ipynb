## Therefore, with copula we aim to model the dependence between the joint variables without necessary putting assumptions on the forms of marginals. In other words, copula can be used to model the dependence of *nd* joint variables where each follows a different distributions (normal, GEV, etc.). 
## in an offical term, a d-dimensional copula, $C[0,1]^{d}:-> [0,1]$ is a **Cumulatie Distribution Function (CDF)** with uniformly distributed marginals.

## $F_{XY} = Pr(X<= x, Y<= y) = C(F_{X}(x), F_{Y}(y))$

# **2. Linear Correlation vs. Rank-based correlations**
## The term correlation is often used to measure the linear dependency between two variables. The famous Pearson correlation metric is the widely used measure of linear dependency. However, in context of copula we are looking to measure the degree of concordance between the joint variables without making any assumption on the form of such association (linear or nonlinear). So the rank-based correlation metric such as Kendall ($\tau$) is often used in the context of copula because 1)it measures the dependence based on ranks i.e., to what extent if one variable increase, the other varible also increases, and 2) it is preserved (invariant) with monotic transformations are applied to the data. Lets see an example:

```python
tau, p_value_MK = stats.kendalltau(data1['Qmax'], data1['Wlmax'])
print('Kendall')
print(tau, p_value_MK)

pr, p_value_Pr= stats.pearsonr(data1['Qmax'], data1['Wlmax'])
print('Pearson')
print(pr, p_value_Pr)
```

```python
data3 = pd.DataFrame().reindex_like(data1)
data3['Qmax'] = np.exp(data1['Qmax'])
data3['Wlmax'] = np.exp(data1['Wlmax'])
data3. head()
```

```python
tau, p_value_MK = stats.kendalltau(data3['Qmax'], data3['Wlmax'])
print('Kendall')
print(tau, p_value_MK)

pr, p_value_Pr= stats.pearsonr(data3['Qmax'], data3['Wlmax'])
print('Pearson')
print(pr, p_value_Pr)

```

Therefore, the ranked-based correlation is more appropriate for measuring the dependency between the dataset than linear correlation metric.

# **4. Empirical copula**

## Empirical copul is the joint CDF function of uniformly distributed marginals and is calculated as :
## $C_{n}(u) = \frac{1}{n} \sum \limits_{i=1}^{n}1(U_{1,i}<= u_{1} , U_{2,i}<= u_{2}),  u = (u_{1},u_{2})$   $\epsilon[0,1]$
## the empirical copula is the most accurate representation of the dependency between the joint dataset. In practice, we want our copula model to fit the empirical copula as close as possible, similar to univariate probability distribution fitting.




## The dependence structure observed by the empirical copula is used to select the most suitable copula among the parametric copulas. Therefore, the numeric goodness of fit metrics of evaluating the copula performance such as Cramer-Von-Mises used in many studies compare the empirical copula with fitted (parameteric) copulas:
## $S_{n} = \sum \limits_{i=1}^{n}(C_{n} - C_{\theta_{n}})^{2}$

# **5. Parameter estimation**
## There are three approaches for estimation of parameters of copulas:
## 1. Parameteric method: parameters of marginal distributions as well as copula function are determined by maximizing a alikelihood function composed of marginal distributions and dependence.

- ## Maximum likelihood:  Parameters of marginal distributions and copula are maximized jointly.
- ## Inference Function for Marginals: First the parameters of marginal distributions are fitted, then these parameters are replaced into the likelihood function whcih is then a function of copula parameters and it is maximized to infer the copula paramters.

## The disadvatage of the parameteric method is that any wrong assumption about the marginal disteribuitions can have significant impact on the propoer estimation of the copula parameters. so it is not recommended <a href = "https://www.sciencedirect.com/science/article/abs/pii/S0167668707001205">Genest et al., (2009)</a>

## 2. Nonparametric method: The parameters of the copula function are determined based on their relationships with the raked-based correlation coefficient (Kendall's $\tau$). This method is useful for copulas with direct relationship between $\tau$ and copula parameters which is not the case for complex copulas and those with more than one parameters (two/three parameters).
## 3. Semi-parametric method: parameters of marginal distributions are determined based on nonparametric (ranked-based) method. Then the pseudo-observations dataset is formed and copula parametrs are determined based on maximizing the pseudo-likelihood function (likelihood function based on pseudo-observations).

# **IMPORTANT: In MhAST, The parameters of the marginal distributions are determined by parameteric method (selecting the best fitted distribution using BIC metric) and then the pseudo-observations are used in a Markov-Chain Monte-Carlo (MCMC) alogirithm to determine the parameters of the copula along with their uncertainty range. The package also calculates the $S_{n}$ metric and its corresponding p-value for all selected copulas. More details will be given in the next section.**






# **6. Bivariate Frequency Analysis using Copula**
## The concept of return period is defined as the average interarrival time between two successive events with probability of exceedance of P:

## $T = \frac{\lambda}{P_{exceedance}}$    
## for example: what is the return period of an event with exeedance probability of 10% every year : T = 1/0.1 = 10 years

## Now in context of multivariate frequency analysis, there are more than one possibility for defining the return period. This is also called "hazard scenarios". For example, we might be interested in finidng the return period of a joint event where the first or second variable exceeds a certain threshold (OR scenario) or a case where both variables excedd the certain threshold (AND). 

 

## $T(X,Y)_{OR} = \frac{\lambda}{P(X>x \ AND \ Y>y)} = \frac{\lambda}{1-C(F_{X}(x), F_{Y}(y))} $

## $T(X,Y)_{AND} = \frac{\lambda}{P(X>x \ AND \ Y>y)} = \frac{\lambda}{1-F_{X}(x)-F_{Y}(y) + C(F_{X}(x), F_{Y}(y))} $
